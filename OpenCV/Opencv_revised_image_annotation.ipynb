```python
ap_img= cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)
print(ap_img.shape)
plt.imshow(ap_img)
```

```python
# Draw a line on that 
ap_img1 = ap_img.copy()
line_img = cv2.line(ap_img1,(400,700),(207,300),(0,255,0),3,cv2.LINE_AA)
plt.figure(figsize=(20,5))
plt.imshow(line_img)
```

```python
# Draw a circle 
ap_img2 = ap_img.copy()
circle_img = cv2.circle(ap_img2,(900,500),22,(255,0,0),3,cv2.LINE_AA)
plt.figure(figsize=(20,5))
plt.imshow(circle_img)
```

```python
# Draw a rectangele
ap_img3 = ap_img.copy()
rectangle_img = cv2.rectangle(ap_img3,(500,100),(700,600),(0,255,255),3,cv2.LINE_AA)
plt.figure(figsize=(20,5))
plt.imshow(rectangle_img)
```

```python
# Put a text on the Image
Titleimg = ap_img.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 1.3
fontFace = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
fontColor = (255, 255, 0)
fontThickness = 2

cv2.putText(Titleimg, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE)
plt.figure(figsize=(20,5))
plt.imshow(Titleimg,cmap="gray")
```
